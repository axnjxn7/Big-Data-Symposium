```python
import dask.dataframe as dd

df = dd.read_csv('2023_combined_mlar_header.txt', 
delimiter = '|',
dtype={'applicant_ethnicity_1': 'float64',
       'applicant_race_1': 'float64',
       'census_tract': 'object',
       'co_applicant_ethnicity_1': 'float64',
       'co_applicant_race_1': 'float64',
       'combined_loan_to_value_ratio': 'object',
       'county_code': 'float64',
       'discount_points': 'object',
       'income': 'float64',
       'interest_rate': 'object',
       'intro_rate_period': 'object',
       'lender_credits': 'object',
       'loan_term': 'object',
       'multifamily_affordable_units': 'object',
       'origination_charges': 'object',
       'prepayment_penalty_term': 'object',
       'property_value': 'object',
       'rate_spread': 'object',
       'total_loan_costs': 'object',
       'total_points_and_fees': 'object',
       'total_units': 'object'})

df.head(3)
```

For this dataset we only want Florida since the competiton is focused on Florida Mortgages. 

```python
df = df[df['state_code'] == 'FL']
df.head(3)
```

Removing columns that have 80% or more NULL, NAs, NaNs. Keep in mind this might take a while to run.

```python
missing_values = df.isnull().mean()

# Compute the missing values
missing_values = missing_values.compute()

# List of columns to remove (80% or more missing values)
columns_removed = missing_values[missing_values >= 0.8].index

# Drop columns where the proportion of missing values is 80% or more
df_cleaned = df.drop(columns=columns_removed)

# Compute the cleaned DataFrame
df_cleaned = df_cleaned.compute()

print(columns_removed)
```

```python
df_cleaned.to_csv('checking.csv', index=False)
```
